```python
import pickle
import plotly.plotly as py
import plotly.graph_objs as go

import plotly
import numpy as np

with open('/tmp/nnet.pkl', 'rb') as f:
    saved = pickle.load(f)
accuracies = saved['accuracies']
groups = saved['groups']
layer_sizes = saved['layer_sizes']
markers = saved['markers']
colours = saved['colours']

scatterClass = go.Scatter
traces = []

unique_groups = np.unique(groups)
for group_idx, group in enumerate(unique_groups):
    ind = np.where(groups==group)
    coords = dict(
        x=layer_sizes[ind].ravel(),
        y=accuracies[ind].ravel()
    )
    thisSymbol = markers[group]
    thisColour = colours[group]
    
    print(group)
    
    trace = scatterClass(
        name=group.strip(),
        mode='markers',
        marker=dict(
            symbol=thisSymbol,
            size=10,
            color=thisColour,
            line=dict(
                width=0.5
            ),
            opacity=1
        ),
        **coords
    )
    traces.append(trace)

layout = go.Layout(
    hovermode='closest',
    title='Best Params NNet',
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)
fig = go.Figure(data=traces, layout=layout)
plot = py.iplot(fig, filename='Best Params NNet')
print('{}'.format(plot.resource))
```
